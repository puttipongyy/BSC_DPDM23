## Christmas and New Year

```python
# Christmas_period
# Define the date period
start_month = 12  # December
start_day = 1
end_month = 12    # December
end_day = 31

# Check if each date is within the specified period every year
data_dadate['Xmask_NY_Period'] = (
    (data_dadate['date_time'].dt.month > start_month) |
    ((data_dadate['date_time'].dt.month == start_month) & (data_dadate['date_time'].dt.day >= start_day))
) & (
    (data_dadate['date_time'].dt.month < end_month) |
    ((data_dadate['date_time'].dt.month == end_month) & (data_dadate['date_time'].dt.day <= end_day))
)

data_dadate
```

```python
data_dadate[data_dadate['Xmask_NY_Period']]
```

## Chinese New Year

```python
# Chinese New Year
# Define the date period
start_month = 1  # January
start_day = 15
end_month = 2    # Febuary
end_day = 15

# Check if each date is within the specified period every year
data_dadate['CNY_Period'] = (
    (data_dadate['date_time'].dt.month > start_month) |
    ((data_dadate['date_time'].dt.month == start_month) & (data_dadate['date_time'].dt.day >= start_day))
) & (
    (data_dadate['date_time'].dt.month < end_month) |
    ((data_dadate['date_time'].dt.month == end_month) & (data_dadate['date_time'].dt.day <= end_day))
)

data_dadate
```

# Function gogo

```python

def support(freqItem):
  return freqItem[1]

def item_description(stockCode, df):
  return set(df[(df['StockCode']==stockCode)]['Description'])


def printFrequentItemsets(freqItem, data, minCombo):
  for i in freqItem:
    if i[0].__len__()>=minCombo:
        print('-'*100)
        print(i[0])
        print('support: ' , i[1]*100)

        for j in i[0]:
          print(j)
          print(item_description(j, data))

        for k in i[2]:
          print('-'*30)
          print(k)
          print('-'*30)
        print('-' * 100)
```
